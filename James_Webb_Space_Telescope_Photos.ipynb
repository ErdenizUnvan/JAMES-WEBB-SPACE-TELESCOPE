```python
#go to search and type images at https://www.jwst.nasa.gov/content/webbLaunch/whereIsWebb.html
#you will find: https://search.usa.gov/search/images?affiliate=jwst&query=Images

#https://search.usa.gov/search/images?affiliate=jwst&query=Images
#https://www.jwst.nasa.gov/content/webbLaunch/whereIsWebb.html
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
url='https://search.usa.gov/search/images?affiliate=jwst&query=Images'
response = requests.get(url,verify=False)
response.text
```

```python
#!pip install bs4
from bs4 import BeautifulSoup
```

```python
#Html parse the source code in order to make it better.
soup = BeautifulSoup(response.content,'html.parser')
soup
```

```python
#once you check the source code you will see that jpg url links will be 
#located at javascript class named class='thumbnail
#so we should get them all
photos = soup.find_all('div',{'class':'thumbnail'})
photos
```

```python
#let us save the file as txt
import io
with open("photos.txt", "w", encoding='utf-8') as file:
    file.write(str(photos))
```

```python
#and open it as a list variable
with open('photos.txt') as f:
    webb=f.read().splitlines()
webb[0]
```

```python
#let us use RE for parsing the data to get src links
import re
bs_images = re.findall('<img[^<>]+>', webb[0])
bs_images
```

```python

for bs_image in bs_images:
    image_path = re.search('src="([^"]+)"', bs_image).group(1)
    #print(image_path)
    with open('webb_photo_links.txt','a') as f:
        f.write(image_path)
```

```python
with open('webb_photo_links.txt') as f:
    photos=f.readlines()
photos
```

```python
  
webb=[]
for photo in photos:
    webb.append(photo.split('jpg'))
webb
```

```python
result=[]
for data in webb:
    print(data)
```

```python
result=[]

for link in data:
    result.append(link+'jpg')
result
```
